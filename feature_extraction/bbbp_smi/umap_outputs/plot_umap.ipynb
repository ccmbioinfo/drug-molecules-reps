```python
import pandas as pd
# import umap
# import umap.plot
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_path = "/hpf/largeprojects/ccmbio/monikas/chem/feature_extraction/bbbp_smi/umap_outputs/"

model1 = "msb-roshan/molgpt"
model4 = "DeepChem/ChemBERTa-77M-MTR"
model5 = "ncfrey/ChemGPT-1.2B"
```

### Model 1: MolGPT

```python
model = model1

def read_csv():
    name = model.split('/')[1]
    filename = input_path + "embeddings_" + name + ".csv"
    # print(f"Filename: {filename}")
    df = pd.read_csv(filename)
    return df

def plot():
    embedding = read_csv()
    # Plot the embedded data points
    _, ax = plt.subplots()
    scatter = ax.scatter(embedding.Dim1, embedding.Dim2, c=embedding.Labels, s=5, alpha=0.5)
    legend1 = ax.legend(*scatter.legend_elements(), title="BBBP")
    ax.add_artist(legend1)
    ax.set_title('BBBP: MolGPT')
    ax.set_xlabel('Dimension 1')
    ax.set_ylabel('Dimension 2')
    plt.show()


plot()

```
