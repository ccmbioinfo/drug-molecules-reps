```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
input_path = "/hpf/largeprojects/ccmbio/monikas/chem/feature_extraction/DUD_2d/DUD/umap_outputs/"

model1 = "msb-roshan/molgpt"
model2 = "gayane/BARTSmiles"
model3 = "zjunlp/MolGen-7b"
model4 = "DeepChem/ChemBERTa-77M-MTR"
model5 = "ncfrey/ChemGPT-1.2B"

target_list = {
               1: 'ace'
               # 2: 'ache',
               # 3: 'ar',
               # 4: 'cdk2',
               # 5: 'cox2',
               # 6: 'dhfr',
            #    7: 'egfr',
               # 8: 'er_agonist',
               # 9: 'fgfr1',
               # 10: 'fxa',
               # 11: 'gpb',
               # 12: 'gr',
               # 13: 'hivrt',
               # 14: 'inha'
            #    15: 'na',
            #    16: 'p38',
            #    17: 'parp',
            #    18: 'pdgfrb',
            #    19: 'sahh',
            #    20: 'src',
            #    21:'vegfr2'
               }

# dot_size = 20
color_active = 'blue'
color_decoy = 'red'
opacity_decoy = 0.3
opacity_active = 0.5
dot_size_decoy = 10
dot_size_active = 50
# alpha = 0.05
marker='+'
```

### Model 1: MolGPT

```python
model = model1
model_name = model.split('/')[1]

def read_csv(target):
    filename = input_path + "embeddings_" + target + "_" + model_name + ".csv"
    df = pd.read_csv(filename)
    return df

def plot():
    
    # Plot the embedded data points
    num_targets = len(target_list)
    ncols = 3
    nrows = (num_targets + ncols - 1) // ncols  # Calculate number of rows needed

    # Create a figure with the specified number of subplots
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, nrows*4))
    axes = axes.flatten()
    red_patch = mpatches.Patch(color=color_active, label='Actives')
    green_patch = mpatches.Patch(color=color_decoy, label='Decoys')
    for axis, target in zip(axes, target_list.values()):
        embedding = read_csv(target)
        print(f"EMbedding: {embedding}")
        colors = [color_active if label == 'actives' else color_decoy for label in embedding.Labels]
        alphas = [opacity_active if label == 'actives' else opacity_decoy for label in embedding.Labels]
        dot_sizes = [dot_size_active if label == 'actives' else dot_size_decoy for label in embedding.Labels]
        axis.scatter(embedding.Dim1, embedding.Dim2, c=colors, s=dot_sizes, alpha=alphas, marker= marker)
        legend1 = axis.legend(handles = [red_patch, green_patch] , title="Target molecules")
        axis.add_artist(legend1)
        axis.set_title(target + ': MolGPT')
        axis.set_xlabel('Dimension 1')
        axis.set_ylabel('Dimension 2')
    
    '''
    # for axis, target in zip(axes, target_list.values()):
    #     embedding = read_csv(target)
    #     print(f"EMbedding: {len(embedding)}")
    #     colors = [1 if label == 'actives' else 0 for label in embedding.Labels]
    #     axis.scatter(embedding.Dim1, embedding.Dim2, c=colors, s=dot_size, alpha=alpha)
    #     legend1 = axis.legend(title="Target molecules")
    #     axis.add_artist(legend1)
    #     axis.set_title(target + ': MolGPT')
    #     axis.set_xlabel('Dimension 1')
    #     axis.set_ylabel('Dimension 2')
    '''


    # Hide any unused subplots
    for i in range(num_targets, len(axes)):
        fig.delaxes(axes[i])
    
    plt.tight_layout()
    plt.show()

plot()

```
